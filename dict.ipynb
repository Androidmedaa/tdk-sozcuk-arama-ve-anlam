şimdi ise diyelim bu dataset uzerinde yani kelimeler uzerinde bir arama yapmak istediğimizde uygun kelimeleri getirsin ve anlamlarını da tıklandıgında getirsin

```python
#bunun için buton ve arama yapması için uygun kelimeleri girebilecegi bir kutucuga ihtiyacim var
#ui için tkinter kutuphanesini kullanicam


import tkinter as tk
from tkinter import scrolledtext
```

```python
import tkinter as tk
from tkinter import scrolledtext

def search_word():
    search_term = entry_search.get().strip()  # Arama terimini al
    result_text.delete('1.0', tk.END)  # Önceki arama sonuçlarını temizle
    
    if not search_term:  # Boş arama yapılmasını engelle
        result_text.insert(tk.END, "Lütfen bir kelime giriniz.")
        return
    
    # Arama işlemini gerçekleştir
    search_result = meaningDF[meaningDF['madde'].str.contains(search_term, case=False)]
    
    # Arama sonuçlarını göster
    if search_result.empty:
        result_text.insert(tk.END, "Aranan kelime bulunamadı.")
    else:
        for index, row in search_result.iterrows():
            result_text.insert(tk.END, f"{row['madde']}: {row['anlam']}\n")
            label = tk.Label(result_text, text="*\n", fg="blue")
            result_text.window_create(tk.END, window=label)
            
def show_meaning(event):
    selected_word = event.widget.get(tk.ACTIVE)
    selected_meaning = meaningDF.loc[meaningDF['madde'] == selected_word, 'anlam'].iloc[0]
    result_text.delete('1.0', tk.END)  # Önceki sonuçları sil
    result_text.insert(tk.END, selected_meaning)
    
    # Geri git düğmesini göster
    button_back.pack(pady=5)

def go_back():
    # Arama sonuçlarını temizle ve geri git düğmesini gizle
    result_text.delete('1.0', tk.END)
    button_back.pack_forget()

# Ana pencereyi oluşturma
root = tk.Tk()
root.title("Sözlük Uygulaması")

# Kelime arama için giriş kutusu ve düğme
frame_search = tk.Frame(root)
frame_search.pack(pady=10)

label_search = tk.Label(frame_search, text="Aranacak Kelime: ")
label_search.grid(row=0, column=0, padx=5, pady=5)

entry_search = tk.Entry(frame_search, width=50)
entry_search.grid(row=0, column=1, padx=5, pady=5)

button_search = tk.Button(frame_search, text="Ara", command=search_word)
button_search.grid(row=0, column=2, padx=5, pady=5)

# Sonuçları göstermek için bir metin kutusu
result_text = scrolledtext.ScrolledText(root, width=80, height=30)
result_text.pack(pady=10)

# Geri git düğmesi
button_back = tk.Button(root, text="Geri Git", command=go_back)

root.mainloop()
```
