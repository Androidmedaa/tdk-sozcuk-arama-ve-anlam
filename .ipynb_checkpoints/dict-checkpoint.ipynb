```python

def show_meaning(event):
    selected_word=event.widget.get(tk.ACTIVE)
    selected_meaning=meaningDF.loc[meaningDF['madde']==selected_word,'anlam'].iloc[0]
    result_text.delete('1.0',tk.END) #onceki sonuclari siler
    result_text.insert(tk.END,selected_meaning)

#creating main window

root=tk.Tk()
root.title("Sozluk Uygulaması")

#kelime listesi
word_list=list(meaningDF['madde']) #dataframeden kelimeleri al
word_listbox=tk.Listbox(root, width=50, height=10)
word_listbox.pack(pady=10)

for word in word_list:
    word_listbox.insert(tk.END, word)
    
word_listbox.bind("<Double-Button-1>", show_meaning) 
#çift tiklama olayini bagla



#sonuclari gostermek için bir metin kutusu
result_text=scrolledtext.ScrolledText(root,width=50,height=50)
result_text.pack(pady=10)

root.mainloop()
    
```
